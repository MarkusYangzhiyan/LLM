## BGD vs. SGD vs. Mini-batch

### 1. BGD

- 全量梯度下降
- 一次吃掉所有数据，算一个平均梯度，然后走一步

```python
for epoch in range(epochs):
    # 假设这里以及定义了 计算梯度的函数 compute_grads
    gradients = compute_grads(X,y,w)
    
    w = w - lr * gradients
```

- 代码中没有内循环，直接全矩阵运算。
- 内存容易爆炸

------------

### 2. SGD

- 随机梯度下降，单个样本
- 一次只吃一个数据，疯狂更新

```python
for epoch in range(epochs):
    # 打乱数据
    indices = np.random.permutation(N)
    X_shuffled, y_shuffled = X[indices],y[indices]

    for i in range(N):
        # 只拿第 i 个样本
        xi = X_shuffled[i : i+1]   
        yi = y_shuffled[i : i+1]

        gradient = compute_grads(xi,yi,w)

        w = w - lr * gradient
        
```

- 必须要写出permutation，否则模型会学会数据的顺序特征，这是一个bias

-----------------

### Mini-Batch

- 工业界标准做法

```python
batch_size = 128

for epoch in range(epochs):
    # 打乱数据，也叫洗牌
    indices = np.random.permutation(N)
    X_shuffled,y_shuffled = X[indices],y[indices]

    for i in range(0,N,batch_size):
        X_batch = X_shuffled[i : i + batch_size]
        y_batch = y_shuffled[i : i + batch_size]

        gradients = compute_grads(X_batch,y_batch,w)

        w = w - lr * gradients
```

- 如果N不能均分batch_size，比如说64，64，....,32，python可以自动处理越界，不会报错
- N = number of samples
- D = number of features(d_model)

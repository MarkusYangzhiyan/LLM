## Qwen2.5-0.5B-Instruct 模型下载部署调用 使用pipeline的简化流程
1. 需要用到的包:
    - modelscope.pipelines -- pipeline
    - modelscope.utils -- Tasks
3. 使用pipe进行模型的加载 model
4. 准备数据messages
5. 调用pipeline加载好的model，返回得到结果
6. 查看结果，输出文本结果


```python
from modelscope.pipelines import pipeline
from modelscope.utils.constant import Tasks
```

```python
model_id = 'Qwen/Qwen2.5-0.5B-Instruct'
```

```python
model = pipeline(
    task = Tasks.chat,
    model = model_id,
    device_map = "auto",
)
```

```python
messages = [
    {'role':'user',"content":'简单陈述一下算法工程师的工作内容和必须要掌握的技术栈和知识是什么？'}
]
```

```python
output = model(
    messages,
    max_new_tokens = 512,
    temperature = 0.7,
    top_p = 0.9
)
```

```python
response = output['message']['content']
print(response)
```
